```python
print("Let's sort the data on the basis of hours in ascending order ")
```

```python
data.sort_values(["Hours"],axis=0,ascending=[True],inplace=True)
data.head(5)
data.plot.bar(x='Hours',y='Scores')
```

From all the above graphs,we can easily conclude that greater the number of hours,greater is the score of the student

Now let's prepare our model

```python
X=data.iloc[:,:-1].values
y=data.iloc[:,1].values
```

We have now divided our data  training and testing purpose

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=0)
```

Training Time

```python
from sklearn.linear_model import LinearRegression
regression=LinearRegression()
regression.fit(X_train,y_train)
print("Training done ")
```

```python
# Plotting the regression line
line = regression.coef_*X+regression.intercept_

# Plotting for the test data
plt.scatter(X, y)
plt.plot(X, line);
plt.show()
```

Testing Time

```python
print(X_test)
print("Prediction of Score")
y_pred=regression.predict(X_test)
print(y_pred)
```

Checking the accuracy

```python
dg=pd.DataFrame({'Actual':y_test,'Predicted':y_pred})
dg
```

```python
hours=[[9.25]]
pred=regression.predict(hours)
print(pred)
```

Model Evaluation

```python
from sklearn import metrics
print('Mean Absolute Error:',metrics.mean_absolute_error(y_test,y_pred))
```
